```python
import pandas as pd
import html5lib
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.investing.com/economic-calendar/philippines-core-cpi-1108'

headers = {'User-Agent': 'Mozilla/5.0'}

def quick_scrape():
    try: 
        response = requests.get(url, headers = headers)
        df_list = pd.read_html(response.text)

        inflation_data = df_list[0]
        print('Scrape Successful!, Here are the first few rows:')
        print(inflation_data.head())

        # Saving it to computer
        inflation_data.to_csv('ph_inflation_scraped.csv', index = False)

    except Exception as e :
        print(f'Error: {e}. Some sites block scraping; manually copy is a choice')

if __name__ == '__main__':
    quick_scrape()
```
